```python
import re
from transliterate import translit



def f(model):
        model = re.sub(r'-'," ", model)
        digit_re = re.compile('\d')
        if digit_re.search(model):
            return [{"name": model, "threshold": 1}]
        if len(model) < 3:
            return [{"name": model, "threshold": 1}, {"name": translit(model, 'ru'), "threshold": 1}]
        if len(model) < 4:
            return [{"name": model, "threshold": 0.97}, {"name": translit(model, 'ru'), "threshold": 0.97}]
        if len(model) < 5:
            return [{"name": model, "threshold": 0.93}, {"name": translit(model, 'ru'), "threshold": 0.93}]
        return [{"name": model, "threshold": 0.85}, {"name": translit(model, 'ru'), "threshold": 0.85}]


brands = list(set(df['Brand_AS'].values))

dictionary = {}

for brand in sorted(brands):
    models = sorted(list(set(df[df['Brand_AS'] == brand]['Model_AS'].values)))
    m = {}
    for model in models:
        id = re.sub(r'-',"", model)
        id = re.sub(r' ',"", id)
        m[id] ={
            "id": id,
            "names":f(model)
        }
    
    id = re.sub(r'-',"", brand)
    id = re.sub(r' ',"", id)
    dictionary[id] = {
        "id": id,
        "names": f(brand),
        "models": m,
    }
```

```python
def replace_rus_to_eng_char(sample_string):
    sample_string = sample_string.upper()
    char_to_replace = {'А': 'A',
                       'В': 'B',
                       'С': 'C',
                       'Е': 'E',
                       'Н': 'H',
                       'К': 'K',
                       'О': 'O',
                       'Р': 'P',
                       'Т': 'T',
                       'Х': 'X'}
    return sample_string.translate(str.maketrans(char_to_replace))
```

```python
def leven(x): 
    words = x['text_brandmodel'].strip().split(' ')
    brand = x['Brand_AS']
    brand_ru = '' if x['Brand_RUS'] is None else x['Brand_RUS']
    for word in words:
        word = replace_rus_to_eng_char(word)
        brand = replace_rus_to_eng_char(brand)
        brand_ru = replace_rus_to_eng_char(brand_ru)
        
        jaro = Levenshtein.jaro_winkler(word, brand)
        levensh = Levenshtein.distance(word, brand)
        length = len(brand)
        if length > 3 and levensh < length and jaro > 0.5 and len(word) <= len(brand):
            return word
        
        jaro = Levenshtein.jaro_winkler(word, brand_ru)
        levensh = Levenshtein.distance(word, brand_ru)
        length = len(brand_ru)
        if length > 3 and levensh < length and jaro > 0.5 and len(word) <= len(brand_ru):
            return word
        
        return np.nan
```

```python
def split_token_by_n(doc, num_token, *pos_str):
    heads = [(doc[num_token], 1) for _ in range(len(pos_str) + 1)]
    points = [0] + list(pos_str), list(pos_str) + [len(doc[num_token].text)]
    words = [doc[num_token].text[prev:curr]  for prev, curr in list(zip(*points))]

    with doc.retokenize() as retokenizer:
        retokenizer.split(
            doc[num_token], 
            words, 
            heads=heads, 
            attrs={}
        )
doc = nlp("I live in NewYork")
split_token_by_n(doc, 3, 1, 2, 3)
[x for x in doc]
```

```python
with jsonlines.open('data/interim/rules/all_rules.jsonl', 'w') as writer:
    writer.write_all(rules)
```
