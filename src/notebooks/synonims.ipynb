```python
import re

def f(model):
        model = re.sub(r'-'," ", model)
        digit_re = re.compile('\d')
        if digit_re.search(model):
            return {"name": model, "threshold": 1}
        if len(model) < 3:
            return {"name": model, "threshold": 1}
        if len(model) < 4:
            return {"name": model, "threshold": 0.97}
        if len(model) < 5:
            return {"name": model, "threshold": 0.93}
        return {"name": model, "threshold": 0.85}
    

brands = list(set(df['Brand_AS'].values))

dictionary = {}

for brand in sorted(brands):
    models = list(set(df[df['Brand_AS'] == brand]['Model_AS'].values))
    models = [f(model)for model in sorted(models)]
    
    id = re.sub(r'-',"", brand)
    id = re.sub(r' ',"", id)
    dictionary[id] = {
        "id": id,
        "names": [f(brand), f(dict[brand])] if brand in dict else [f(brand)],
        "models": models,
    }
```

```python
def replace_rus_to_eng_char(sample_string):
    sample_string = sample_string.upper()
    char_to_replace = {'А': 'A',
                       'В': 'B',
                       'С': 'C',
                       'Е': 'E',
                       'Н': 'H',
                       'К': 'K',
                       'О': 'O',
                       'Р': 'P',
                       'Т': 'T',
                       'Х': 'X'}
    return sample_string.translate(str.maketrans(char_to_replace))
```
